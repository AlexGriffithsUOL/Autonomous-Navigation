Converting all images in the folder to PNG from PGM

```python
#Conversion
import os
from PIL import Image

for file in os.listdir():
    filename, extension = os.path.splitext(file)
    if extension == ".pgm":
        new_file = "{}.jpg".format(filename)
        with Image.open(file) as im:
            im.save(new_file)
```

Isolating the wall using opencv

```python
import cv2
import numpy as np
testimg = cv2.imread("MapUnlearnt.jpg")
grayscaled = cv2.cvtColor(testimg,cv2.COLOR_BGR2GRAY)
retval, threshold = cv2.threshold(grayscaled, 100, 255, cv2.THRESH_BINARY_INV)
retval2, threshold2 = cv2.threshold(grayscaled, 230, 255, cv2.THRESH_BINARY_INV)
```

```python
plt.figure()
f, axarr = plt.subplots(1,2)
axarr[0].imshow(threshold, cmap="gray")
axarr[1].imshow(threshold2, cmap="gray")
```

```python
dstHolder = cv2.Mat(threshold2)
maskHolder = cv2.Mat(threshold2)
subtracted = cv2.subtract(threshold2, threshold, dstHolder, maskHolder, -1)
```
